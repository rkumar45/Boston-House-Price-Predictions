**Modelling**

```
# splitting up the dataset into train and test datasets

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size = 0.4, random_state = 44)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```
# PCA 

from sklearn.decomposition import PCA

pca = PCA(n_components = None)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
explained_variance = pca.explained_variance_ratio_

print(explained_variance)
```

```
pca = PCA(n_components = 1)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
explained_variance = pca.explained_variance_ratio_

print(explained_variance)
```

**Linear Regression**

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


# creating a linear regression model to predict the values
model = LinearRegression()

# fitting the training set into the model
model.fit(x_train, y_train)


# checking the model coefficients
model.coef_
```

```
# predicting the values and storing it in a variable named 'y_pred'

y_pred = model.predict(x_test)

print(y_pred)
```

```
# printing the values of y_test so that we can compare the real and predicted values and check the error

y_test
```

*Ridge Regression* *(L2 regularization)*

```
from sklearn.linear_model import Ridge

# creating a model
model = Ridge()

# fitting the training set into the model
model.fit(x_train, y_train)

# predicting the results for test set
y_pred = model.predict(x_test)

# calculating the mean squared error
mse = np.mean((y_pred - y_test)**2)
print("MSE :",mse)

# calculating the root mean squared error
rmse = np.sqrt(mse)
print("RMSE :", rmse)

# calculating the r2 score 
r2 = r2_score(y_pred, y_test)
print("r2 score : %.2f" % r2)
```

*Lasso Regression* *(L1 regularization)*
